```python
import requests
```

```python
# API parameters
api_key = '2bfadace9e23a0c08c0027cdf60fd2a2'
latitude = -37.92512
longitude = 145.11429
units = 'metric'

# Construct the URL
base_url = "https://api.openweathermap.org/data/2.5/forecast"
url = f"{base_url}?lat={latitude}&lon={longitude}&appid={api_key}&units={units}"

# Make the API call
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    data = response.json()
    print(data)
else:
    print(f"Error: {response.status_code}")
    print(response.text)
```

```python
import json
from datetime import datetime

# Assuming the JSON response is stored in a variable called 'json_response'
# For this example, I'll use the first part of the JSON you provided

json_response = {'cod': '200', 'message': 0, 'cnt': 40, 'list': [{'dt': 1720764000, 'main': {'temp': 285.91, 'feels_like': 285.25, 'temp_min': 285.18, 'temp_max': 285.91, 'pressure': 1015, 'sea_level': 1015, 'grnd_level': 1006, 'humidity': 77, 'temp_kf': 0.73}, 'weather': [{'id': 803, 'main': 'Clouds', 'description': 'broken clouds', 'icon': '04d'}], 'clouds': {'all': 52}, 'wind': {'speed': 2.5, 'deg': 15, 'gust': 3.62}, 'visibility': 10000, 'pop': 0.17, 'sys': {'pod': 'd'}, 'dt_txt': '2024-07-12 06:00:00'}]}  # The full JSON response would go here

def kelvin_to_celsius(kelvin):
    return round(kelvin - 273.15, 2)

def format_forecast(forecast):
    return {
        'date': datetime.fromtimestamp(forecast['dt']).strftime('%Y-%m-%d %H:%M:%S'),
        'temperature': kelvin_to_celsius(forecast['main']['temp']),
        'feels_like': kelvin_to_celsius(forecast['main']['feels_like']),
        'humidity': forecast['main']['humidity'],
        'weather': forecast['weather'][0]['main'],
        'description': forecast['weather'][0]['description'],
        'wind_speed': forecast['wind']['speed'],
        'wind_direction': forecast['wind']['deg'],
        'visibility': forecast['visibility'],
        'precipitation_probability': forecast['pop']
    }

readable_forecasts = [format_forecast(forecast) for forecast in json_response['list']]

print(json.dumps(readable_forecasts, indent=2))
```
